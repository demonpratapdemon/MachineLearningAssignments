## Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web application framework written in Python. It is classified as a microframework because it does not require particular tools or libraries. It has no database abstraction layer, form validation, or any other components where pre-existing third-party libraries provide common functions.

Advantages of Flask Framework are as follows:

1. Flask is a lightweight framework with minimalist features.
2. Flask is suitable for single-page application.
3. Flask provides an in-built debugger that offers virtual debugging
4. Flask offers flexibility to the developers as it is a micro-based framework with extensible libraries
5. Flask offers in-built security features for common threads like CSRF, SQL Injection XSS, etc

## Q3. What is App routing in Flask? Why do we use app routes?

App routing is the technique used to map the specific URL with the associated function intended to perform some task. The Flask Web frameworks use the routing technique to help users remember application URLs.

The app routes are used to map to different functions and their necessary functionality based on a particular URL. A web application can be doing multiple independent things and as a result of their disjoint behaviour we have different methods and for accessing each of these methods from outside world, we need to map different URLs to these methods.

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

To build a URL to a specific function, use the url_for() function. It accepts the name of the function as its first argument and any number of keyword arguments, each corresponding to a variable part of the URL rule.

```python
from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route("/student")
def student():
    return "student"

@app.route("/college")
def college():
    return "college"

@app.route("/faculty")
def faculty():
    return "faculty"

@app.route("/univ/<name>")
def univ(name):
    print(name)
    if name == "student":
        return redirect(url_for("student"))
    if name == "college":
        return redirect(url_for("college"))
    if name == "faculty":
        return redirect(url_for("faculty"))

if __name__ == "__main__":
    app.run(host= "0.0.0.0", port = 8081)
```

Even we hit the url http://192.168.29.146:8081/univ/student, then also it will redirect to the url http://192.168.29.146:8081/student to the appropriate route and the necessary fucntion will be executed.
