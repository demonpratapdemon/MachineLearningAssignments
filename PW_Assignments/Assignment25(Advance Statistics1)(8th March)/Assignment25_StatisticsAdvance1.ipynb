## Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

The properties of the binomial distribution are:

1. There are two possible outcomes: true or false, success or failure, yes or no.
2. There is n number of independent trials or a fixed number of n times repeated trials.
3. The probability of success or failure remains the same for each trial.
4. Only the number of success is calculated out of n independent trials.
5. Every trial is an independent trial, which means the outcome of one trial does not affect the outcome of another trial.

Examples of 2 events where Bonomial Distribution can be applied are as follows:

1. By using the YES/ NO survey, we can check whether the number of persons views the particular channel.
2. Taking a survey of positive and negative reviews from the public for any specific product or place.
3. Tossing of a coin "n" number of times

## Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

The Poisson cumulative distribution function lets us obtain the probability of an event occurring within a given time or space interval less than or equal to x times if on average the event occurs  times within that interval.



```python
# Here mean is same as lambda(lamb)
def poisson_cdf(lamb, point):
    cdf = 0
    for i in range(point + 1):
        cdf = cdf + math.exp(-lamb) * (math.pow(lamb, i) / math.factorial(i))
    return cdf
```

```python
point = 4
mean = 1
cdf = poisson_cdf(mean, point)
print("CDF of a Poisson Distribution at a given point = ", cdf)
```

Verifying with the Scipy Poisson CDF

```python
scipy_poisson_cdf = stats.poisson.cdf(4, mu = 1)
print("Scipy Poisson CDF = ", scipy_poisson_cdf)
```

|**Distribution**| The Binomial Distribution is most of the times follow Gaussian Distribution| The Poisson Distribution is asymmetric; it is always skewed toward the right.|

|**Parameter**|n = no of events, p = probaility of success| $\lambda$ = expected no of events to occur at a time interval|
