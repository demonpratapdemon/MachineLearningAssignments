## Q1. Explain GET and POST methods.

There are various protocols defined by HTTP in order to make communication between client and server possible. The GET and POST are 2 such protocols mentioned by HTTP.

**GET** - GET is used to request data from a specified resource. The query string (name/value pairs) is sent in the URL of a GET request. GET requests remain in the browser history and it has a length restriction. This method should not be used when we are dealing with sensitive data.

**POST** - POST is used to send data to a server to create/update a resource. The data sent to the server with POST is stored in the request body of the HTTP request. POST requests are never cached and it doesn't have a length restriction. The data sent to the server is in the JSON format.

## Q2. Why is Request used in Flask?

In a client and server architecture, there is always a flow of data from both the sides. Whenever the client sends the data, it has to be consumed by the server in order to process the exact information required. The **Request** in Flask, is an object that contains all the data sent from the Client to Server. This data can be recovered using the GET/POST Methods. The variuous methods for Request are as follows:
1. **request.args.get("name")** - For getting the value from a query param in GET request.
2. **request.form.get("form_key_name")** - For getting the value from a POST request body.
3. **request.get_json()** - This converts the JSON object into Python data.

## Q3. Why is redirect() used in Flask?

Flask redirect is defined as a function or utility in Flask which allows developers to redirect users to a specified URL and assign a specified status code. A redirect is used in the Flask class to send the user to a particular URL with the status code. conversely, this status code additionally identifies the issue. When this function is called, a response object is returned, and the redirection happens to the target location with the status code. Parameters:

1. **location(str):** The location which URL directs to.
2. **code(int):** The status code for Redirect.
3. **Code:** The default code is 302 which means that the move is only temporary.

Flask also facilitates us to return the response in the form of HTML templates in addition to return response of string or JSON or any other format.
