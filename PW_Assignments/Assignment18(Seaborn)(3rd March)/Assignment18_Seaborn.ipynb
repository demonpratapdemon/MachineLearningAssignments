## Q1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

The 5 different plots that can be plotted using Seaborn Library are as follows along with their use:

**1. Stripplot** - In .stripplot() function we have to define which attribute of the dataset to be on the x-axis and which attribute of the dataset should on y-axis.data = iris means attributes which we define earlier should be taken from the given data. It helps in undertsanding the spread of data along the y-axis.

**2. Barplot** - A barplot is basically used to aggregate the categorical data according to some methods and by default it is the mean. It can also be understood as a visualization of the group by action. To use this plot we choose a categorical column for the x-axis and a numerical column for the y-axis, and we see that it creates a plot taking a mean per categorical column.

**3. Countplot** - A countplot basically counts the categories and returns a count of their occurrences. It is one of the simplest plots provided by the seaborn library.

**4. Boxplot** - Box Plot is the visual representation of the depicting groups of numerical data through their quartiles. Boxplot is also used to detect the outlier in the data set

**5. Violinplot** - It is similar to the boxplot except that it provides a higher, more advanced visualization and uses the kernel density estimation to give a better description about the data distribution.

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
df = sns.load_dataset("fmri")
df.head()
```

```python
# Find the unique regions and events in the dataframe
print(df["region"].unique())
print(df["event"].unique())
```

```python
sns.lineplot(data= df, x = "timepoint", y = "signal", hue  = "region", style= "event")
plt.title("TImepoint vs Signal in different Regions for different Events")
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

## Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic_df = sns.load_dataset("titanic")
titanic_df.head()
```

```python
sns.boxplot(data=titanic_df, x = "pclass", y = "age")
```

```python
sns.boxplot(data=titanic_df, x = "pclass", y = "fare")
```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds_df = sns.load_dataset("diamonds")
diamonds_df.head()
```

```python
diamonds_df["cut"].unique()
```

```python
sns.histplot(data=diamonds_df, x = "price", hue = "cut", kde = True)
plt.title("Distribution of Diamonds Price for Different Cuts")
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris_df = sns.load_dataset("iris")
iris_df.head()
```

```python
sns.pairplot(data= iris_df, hue = "species")
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights_df = sns.load_dataset("flights")
flights_df.head()
```

```python
flights_mat = flights_df.pivot(index="year", columns="month", values= "passengers")
flights_mat
```

```python
sns.heatmap(data = flights_mat, annot = True, linewidths=0.5, cmap="coolwarm", fmt = 'd')
```
