## Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.

Examples of situation where the Normal Distribution can be used for modelling are as follows:

1. Height, Weight, Age of people
2. Rolling a die
3. Tossing a coin
4. IQ

The two main parameters of a (normal) distribution are the mean and standard deviation. The parameters determine the shape and probabilities of the distribution. The shape of the distribution changes as the parameter values change.

The mean is used by researchers as a measure of central tendency. It can be used to describe the distribution of variables measured as ratios or intervals. In a normal distribution graph, the mean defines the location of the peak, and most of the data points are clustered around the mean.

The standard deviation measures the dispersion of the data points relative to the mean. It determines how far away from the mean the data points are positioned and represents the distance between the mean and the observations.

The Normal Distribution is symmetrical arounf the mean. Normal Distrubiton follows 68-95-97 rule which means:

1. 68% of data lies within the first standard deviation
2. 95% of data lies within the second standard deviation
3. 97% of data lies within the third standard deviation

## Q5: What is Bernoulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

In probability theory and statistics, the Bernoulli distribution, named after Swiss mathematician Jacob Bernoulli, is the discrete probability distribution of a random variable which takes the value 1 with probability p and the value 0 with probability ${q=1-p}$. The experiments under consideration for Bernoulli Distribution exactly have 2 outcomes - true/ success and false/failure.

Examples of Bernoulli Distribution:

1. Winning a match
2. Tossing a coin one time

The difference between Bernoulli and Binomial Distributions are as follows:

1. Bernoulli deals with the outcome of the single trial of the event, whereas Binomial deals with the outcome of the multiple trials of the single event.
2. Bernoulli is used when the outcome of an event is required for only one time, whereas the Binomial is used when the outcome of an event is required multiple times.

## Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.

We have a dataset which is Normally Distributed and has a mean($\mu$) = 50 and standard deviation($\sigma$)=10.

We have to find the probability when a randomly selected observation has a value greater than 60. For this purpose, we are going to leverage the use of Z-Score and Z-table. The formular for Z-Score is as follows:

$$ Z = (X - \mu) / \sigma$$

Here X = observed value,
$\mu$ = mean,
$\sigma$ = standard deviation

The Z-Score helps to understand the probability of getting the observed value from the left hand side.

Let's calculate the Z-Score for the values:

$$Z = (60 - 50) / 10 = 1 $$

So the observed Z-Score is positive. We take a look at the Z-table from the below mentioned  link and get the value:

<a href="https://www.math.arizona.edu/~rsims/ma464/standardnormaltable.pdf">Z-table PDF</a>

So the observed Z-Score from the above table = .84134
 But since we need the probability of the observed value to be greater than 60 we have to deduct this from 1 hence the required value = $1 - .84134 = 0.15866$
 
 Verifying the same with Scipy

```python
obs = 60
mean = 50
std = 10
z_score = (obs - mean) / std

prob = 1 - stats.norm.cdf(z_score)
print("Probability of getting a value of 60 with the above mentioned Normal Distribution =", prob)
```

## Q8: What is the Z score? State the importance of the Z score.

Z-score is a statistical measurement that describes a value's relationship to the mean of a group of values. Z-score is measured in terms of standard deviations from the mean. If a Z-score is 0, it indicates that the data point's score is identical to the mean score. A Z-score of 1.0 would indicate a value that is one standard deviation from the mean. Z-scores may be positive or negative, with a positive value indicating the score is above the mean and a negative score indicating it is below the mean.

The statistical formula for a value's z-score is calculated using the following formula:

$$z = ( x - μ ) / σ $$

A Z-score is important because it tells where the data lies in the data distribution. For example, if a Z-score is 1.5, it is 1.5 standard deviations away from the mean. Because 68% of the data lies within one standard deviation (if it is normally distributed), 1.5 might be considered too far from average for your comfort.

## Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

The Central Limit Theorem(CLT) says that the sampling distribution of the mean of a larger population of data will always be Normally Distributed considering the sample size is long enough. Whether the distribution is Norma, Poisson, Bernoulli or any other distribution, the distrubiton of the mean of the sampling data will always be Normally distributed.

The significance of CLT are as follows:

1. The central limit theorem is useful when analyzing large data sets because it allows one to assume that the sampling distribution of the mean will be normally-distributed in most cases. This allows for easier statistical analysis and inference.
2. The features may be widely distributed in different scales and plotting them will scatter the data points. But CLT will help in bringing to same level as much as possible.

## Q10: State the assumptions of the Central Limit Theorem.

The assumptions of Central Limit Theorem are as follows:

1. The data must adhere to the randomization rule. It needs to be sampled at random.
2. The samples should be unrelated to one another. One sample should not impact the others.
3. When taking samples without replacement, the sample size should not exceed 10% of the population.
