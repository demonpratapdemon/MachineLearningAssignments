## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np  
np.random.seed(3)  
x = 3 + np.random.normal(0, 2, 50)  
y = 3 + np.random.normal(0, 2, len(x))  
Note: Also add title, xlabel, and ylabel to the plot.

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
print(x)
print(y)
```

```python
plt.scatter(x = x, y = y, c = "g", marker = "X", alpha = 1)
plt.xlabel("x data")
plt.ylabel("y data")
plt.title("x vs y")
plt.grid()
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

import numpy as np  
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])  
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])  
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])  
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The **subplot** function adds subplot to a current figure at the specified grid position. It is similar to the subplots() function however unlike subplots() it adds one subplot at a time. So to create multiple plots you will need several lines of code with the subplot() function. Another drawback of the subplot function is that it deletes the preexisting plot on your figure.

```python
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
ax[0,0].plot(x, y1)
ax[0,1].plot(x, y2)
ax[1,0].plot(x, y3)
ax[1,1].plot(x, y4)
ax[0,0].set_title("Line 1")
ax[0,1].set_title("Line 2")
ax[1,0].set_title("Line 3")
ax[1,1].set_title("Line 4")
ax[0,0].set_ylabel("y1")
ax[0,1].set_ylabel("y2")
ax[1,0].set_ylabel("y3")
ax[1,1].set_ylabel("y4")
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. 

A bar chart is used to find the comparisons between the discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(x = company, height = profit)
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)  
box2 = np.random.normal(90, 20, 200)

A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

&nbsp;&nbsp;**Minimum value :** The smallest observation in the dataset.  
&nbsp;&nbsp;**First quartile (Q1) :** The value that separates the lowest 25% of the dataset from the rest.  
&nbsp;&nbsp;**Median (Q2) :** The value that separates the lowest 50% of the dataset from the highest 50%.  
&nbsp;&nbsp;**Third quartile (Q3) :** The value that separates the lowest 75% of the dataset from the rest.  
&nbsp;&nbsp;**Maximum value :** The largest observation in the dataset.

Box plots are commonly used to visualize the distribution of a dataset and to identify potential outliers or extreme values. They are particularly useful for:

**Summarizing the distribution of a dataset :** Box plots provide a visual summary of the distribution of a dataset based on its five-number summary (minimum, first quartile, median, third quartile, and maximum). This allows us to quickly understand the range, central tendency, and spread of the data.

**Identifying potential outliers :** Box plots can also be used to identify potential outliers or extreme values in a dataset. Any data points that fall outside of the whiskers (i.e., more than 1.5 times the interquartile range from the nearest quartile) are considered outliers and are plotted as individual points. This allows us to quickly identify any unusual or extreme values in the data.

**Comparing distributions :** Box plots can also be used to compare the distribution of two or more datasets. By plotting multiple box plots side by side, we can compare the range, central tendency, and spread of each dataset and identify any differences or similarities between them.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data = [box1, box2]
fig, ax = plt.subplots()
ax.boxplot(data)
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
ax.set_title('Box Plot of Two Combined Datasets')
plt.show()
```
